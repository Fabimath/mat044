# Modelos de Regresión

```r
pagina = 'https://raw.githubusercontent.com/Fabimath/mat044/master/ay3/data.csv'
x <- url(pagina, "r")
x <- readLines(x)
data = read.csv(text = x)
```

```r
head(data)
```

```r
plot(data$x,data$y)
```

Recordemos que si pensamos que $y_i = \beta_0 + \beta_1 * x_i + error$ entonces el estimador de minimos cuadrados para $\beta_0$ y $\beta_1$ vienen dados por:
$$
\widehat{\beta_1} = \dfrac{S_{xy}}{S_{xx}} \quad , \quad \beta_0 = \overline{y} - \widehat{\beta_1}\cdot \overline{x}
$$

Nota que:

```r
n = length(data$x)
sxy = sum(data$x*data$y) - n*mean(data$x)*mean(data$y)
sxx = sum(data$x^2) - n*mean(data$x)^2
b1 = sxy/sxx
b0 = mean(data$y) -b1*mean(data$x)
```

```r
print(b0)
print(b1)
```

```r
#genero una banda de predicciones
pp = seq(-5,5,0.1)
predict = b0*rep(1,length(pp)) + b1*pp
```

```r
plot(data)
points(pp,predict,col='red')
```

```r
# en el R
reg = lm(y~x,data)
plot(data)
abline(reg)
```

```r
R2 = (cov(data$x,data$y)/(sd(data$x)*sd(data$y)))^2
R2
```
